```python
from pager import RegionModel, RowsModel, PDFModel, PDF2Rows, Rows2Regions
```

```python
pdf = PDFModel()
pdf2rows = PDF2Rows()
rows = RowsModel()
rows2region = Rows2Regions()
regions = RegionModel()
```

```python
pdf.read_from_file('../tests/files/text_header_table.pdf', method='r')
```

```python
pdf2rows.convert(pdf, rows)
```

```python
rows2region.convert(rows, regions)
```

```python
for r in regions.regions:
    print(r.text)
```
