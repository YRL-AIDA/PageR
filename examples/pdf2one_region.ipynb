```python
from pager import PDFModel, PDF2Words, WordsModel, RegionModel, WordsToOneBlock, ImageModel, PDF2Img
```

```python
pdf = PDFModel()
pdf2words = PDF2Words()
words = WordsModel()
words2regions = WordsToOneBlock()
regions = RegionModel()
img = ImageModel()
pdf2img = PDF2Img()
```

```python
pdf.read_from_file('../tests/files/text_header_table.pdf', method='w')
pdf2words.convert(pdf, words)
words2regions.convert(words, regions)
```

```python
pdf2img.convert(pdf, img)
```

```python
img.show()
for r in regions.regions:
    r.segment.plot()
```
