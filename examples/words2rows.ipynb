```python
from pager import ImageModel, WordsModel, RowsModel, RegionModel, Image2Words, Words2Rows, Rows2Regions , PDF2Img, PDFModel, PDF2Words
```

```python
pdf = PDFModel()
img = ImageModel()
words = WordsModel()
rows = RowsModel()
regions = RegionModel()

pdf2img = PDF2Img()
pdf2words = PDF2Words()
img2words = Image2Words()
words2rows = Words2Rows()
rows2regions = Rows2Regions()
```

```python
pdf.read_from_file('../tests/files/text_header_table.pdf', 'w')
pdf2img.convert(pdf, img)
img.show()
```

```python
pdf2words.convert(pdf, words)
words2rows.convert(words, rows)
rows2regions.convert(rows, regions)
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi=300)
img.show()

for w in words.words:
    w.segment.plot(color='r', width=0.2)
for row in rows.rows:
    row.segment.plot(color='b', width=0.4)
for reg in regions.regions:
    reg.segment.plot(color='orange')
```
